# 1. Linear Regression
This notebook illustrates a couple of different packages you can use to do a simple linear regression in Python. 


## Aim
Explore a relatively popular and simple dataset to explore statistical packages for linear regression. I'm going to use the USA Housing dataset to have a look at how the different packages work, and my goal is try and build this notebook up as a simple cheatsheet for linear regression in Python.


```python
#imports
import pandas as pd
import numpy as np

#visualisation
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

#training and testing sets
from sklearn.model_selection import train_test_split

#statsmodel
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.tools.eval_measures import rmse

#sklearn 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```


## The Data 
USA Housing data from Kaggle (https://www.kaggle.com/datasets/vedavyasv/usa-housing). There's no copyright for this data so feel free to explore it yourself if you're interested. 

```python
dataset = pd.read_csv('USA_Housing.csv')
```

Let's visualise the plots so we can have a look at the dataset and see which two variables show a linear trend. Seaborn's pairplot is a pretty great tool to use here.

## Data Wrangling 

Our next step is to clean up the dataset. For linear regression, we need to have continuous and nuerical datasets. We are interested in only the numerical columns, so let's drop the Address column. 

```python
dataset.drop('Address', axis = 1, inplace = True)
```

```python
dataset.info()
```

```python
dataset.shape
```

### Missing Data? 
We need to confirm there's no data points missing. If there were features where data points were missing or non-numerical, we'd have to think of a way of handling that before we can do any sort of regression analysis. We are pretty lucky here that the dataset has no data missing!

```python
dataset.isnull().sum()
```

This is a personal preference but I like renaming the columns as spaces make things just more annoying to call later on. 

```python
dataset = dataset.rename(columns = {'Avg. Area Income' : 'Avg_Area_Income', 
                                    'Avg. Area House Age': 'Avg_Area_House_Age', 
                                    'Avg. Area Number of Rooms': 'Avg_Area_Number_of_Rooms',
                                    'Avg. Area Number of Bedrooms': 'Area_Population'})
dataset.columns
```

## Simple Linear Regression: Average Area Income and House Prices

In he pair plot there were somewhat linear trends between quite a few varaibles and house prices. I'm going to focus on Average Area Income how this affects house prices. 

```python
ax = sns.scatterplot(x='Avg_Area_Income', y = 'Price', data = dataset)
```

We can almost see the straight line through the middle of the datapoints. It's worth noting there are some stragglers from the general trend: for example we can see one data point with an Average Area Income of USD 20,000 stands out quite distinctively from our dataset. This is most likely an outlier.

# Data Processing

I am going to split up our data here so build the model using only 80% of the dataset, the training set. In order to judge the quality of this model in predicting house prices, the remaining 20% of the dataset will be the testing set. 

```python
X = dataset[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms',
       'Area_Population', 'Area Population']]
y = dataset['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, random_state= 101)
print(X_train.shape, y_train.shape)
```

```python
print(X_test.shape, y_test.shape)
```

The training set has 4000 rows and the testing set 1000. 

### Generating our Linear Regression Models and Analysis
There's quite a few different ways of doing this, I'll focus on Scikit Learn, Statsmodels, and Pingouin. 


#### Scikit-Learn
Pros: Scikit Learn is optimised for preditions 

```python
feature_train = X_train[['Avg_Area_Income']].values
feature_test = X_test[['Avg_Area_Income']].values
simple_linear_reg = LinearRegression()
simple_linear_reg.fit(feature_train, y_train)
y_pred = simple_linear_reg.predict(feature_test)
print("Pearson's Correlation Coefficient, r ", np.sqrt(metrics.r2_score(y_test, y_pred)))
print('R-Squared', metrics.r2_score(y_test, y_pred))
print('Mean Absolute Error',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Square Error', metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square Error', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

A pearson's correlation coeffient of 0.64 suggests there is a moderate positive linear correlation between the average area income and house prices. 

```python
sns.regplot(x= y_pred, y = y_test, line_kws={'color':'red'})
plt.ylabel('Actual Price')
plt.xlabel('Predicted Price')
plt.title('Acual Price vs Predicted Price')
```

```python
print('Coefficient: ', simple_linear_reg.coef_)
print('Intercept: ', simple_linear_reg.intercept_)
```

Residuals 

```python
residuals = (y_test - y_pred)
sns.regplot(x= y_pred, y = residuals, line_kws={'color':'red'})
plt.xlabel('Actual Price')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Price')
```

The residuals look pretty symmetrically distributed, scattered closely towards the centre. There doesn't seem to be an obvious patern either, so we can conclude that a linear regression is appropriate, and the association is moderate.

#### Statsmodels
Pros: Optimised for insight

```python
training_set = pd.concat([X_train, y_train], axis = 1)
training_set.head()
```

```python
linear_model_stats = ols("Price ~ Avg_Area_Income", data = training_set).fit()
print(linear_model_stats.summary())
```

```python
fig = plt.figure(figsize=(14,6))
fig = sm.graphics.plot_regress_exog(linear_model_stats, 
                                       'Avg_Area_Income',
                                       fig=fig)
```

Statsmodel is much faster and more convenient at providing metrics on the data- only one line can help us look at quite a few different types of analysis tools to get deeper understanding of the data. 
But, what if we wanted to perform the same analysis as we did with Scikit Learn? The equation of the line for both methods is the same as both the packages are producing a least squares regression line. So, let's make a testing set for our data. 

```python
testing_set = pd.concat([X_test, y_test], axis = 1)
testing_set.head()
```

```python
y_pred = linear_model_stats.predict(X_test)
```

```python
testing_set = pd.concat([testing_set, y_pred], axis = 1)
testing_set.rename(columns={0:'Predicted_Price'}, inplace=True)
```

```python
testing_set
```

```python
sns.regplot(x= 'Predicted_Price', y = 'Price', data = testing_set, line_kws={'color':'red'})
plt.ylabel('Actual Price')
plt.xlabel('Predicted Price')
plt.title('Acual Price vs Predicted Price')
```

It's worth noting that seaborn also has its own built in residual plot function. 

```python
sns.residplot(x= 'Predicted_Price', y = 'Price', data = testing_set)
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted Price')
```

## Scikit Learn vs. Statsmodel?
It's obvious they both are very powerful packages. When I'm producing a report about a dataset, So, when should we use what, or more importantly, when is one more useful than the other. Statsmodel is designed to provide insight into datasets- that doens't mean it can't make predictions- it's just better and easier to use when you want to study the data in more detail. Scikit Learn, on the other hand, is much better suited for predictions.
